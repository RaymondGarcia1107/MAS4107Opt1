# Testing Gauss Seidel Loss Functions
This notebook tests the infinite norm and l2 norm for the Gauss Seidel method. The goal is to understand the difference between each norm and choose one going forward. Generally, the infinite(max) norm will be stricter than the l2 norm. If the difference in steps to converge are negligible, we should opt to use the infinite norm. This notebook will aim to answer this question on a 50x50 matrix

```python
from HeatModel import *
```

Initializing a HeatModel A of size 50

```python
n = 50
A = HeatModel(n)
```

Running the GS method without stopping

```python
x1, count, tols = A.GaussSeidel(nIter=9000, stop = False)
```

For a 50 x 50 matrix, 9,000 iterations on an intel i7-13700H in an XPS 15 chassis takes about 1 minute and 30 seconds.

## Results
Now to plot the results in a log scaled graph

```python
plt.semilogy(tols[0], label = f"maxNorm: {tols[0,-1]:.9f}")
plt.semilogy(tols[1], label = f"l2Norm: {tols[1,-1]:.9f}")
plt.legend();
```

Here we can see that both norms are very close. It is impossible to see the difference between the two

```python
maxNormCount = np.where(tols[0] < 1e-8)
print(f"The max norm reached 1e-8 in {maxNormCount[0][0]} iterations")

l2NormCounts = np.where(tols[1] < 1e-8)
print(f"The l2 norm reached 1e-8 in {l2NormCounts[0][0]} iterations")
```

## Final Verdict
The max norm reached tolerance in 6,297 iterations, while l2 reached tolerance in 6,236. This difference is negligible. Given that max norm is stricter, we should use the max norm for convergence.
